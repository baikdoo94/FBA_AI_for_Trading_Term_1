가끔 오류가 발생하여 plotly, tqdm를 추가적으로 설치 해줘야 한다. 

plotly 의 경우 오류가 종종 발생 하는데, 

        1] 설치가 되어 있지 않다면, 관리자로 cmd 를 실행시킨 다음에   pip install plotly==2.2.3
        2] 설치가 이미 되어 있다면, 관리자로 cmd 를 실행시킨 다음에   pip install plotly==2.2.3 --force-reinstall
